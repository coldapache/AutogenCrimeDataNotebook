```python
import pandas as pd
import numpy as np

# generate fake data for crime type, latitude, and longitude
crime_type = np.random.choice(['theft', 'assault', 'burglary', 'vandalism', 'robbery'], size=100)

#creates the random latitude values
lat = np.random.uniform(low=30.0, high=50.0, size=100)

#creates the random longitude values
lon = np.random.uniform(low=-120.0, high=-60.0, size=100)

# creates a DataFrame with columns for our fake crime data
df = pd.DataFrame({'crime_type': crime_type, 'latitude': lat, 'longitude': lon})

# view the DataFrame
print(df)
```

```python
import matplotlib.pyplot as plt

#counts the rows of specific crime types and sums them 
counts = df['crime_type'].value_counts()

#builds and visualizes the actual pie chart
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')


# apply styling
plt.style.use('dark_background')
plt.rcParams['font.size'] = 16
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['text.color'] = 'white'

#outputs the pie chart
plt.axis('equal')
plt.show()
```
